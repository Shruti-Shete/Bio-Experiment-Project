```python
import pandas as pd
import numpy as np
```

```python
M_24 = pd.read_csv("C:\\Users\\DELL\\Desktop\\Shree_Tableau_files\\Myeloid_24h_v3.csv")
M_48 = pd.read_csv("C:\\Users\\DELL\\Desktop\\Shree_Tableau_files\\Myeloid_48h_v3.csv")
M_72 = pd.read_csv("C:\\Users\\DELL\\Desktop\\Shree_Tableau_files\\Myeloid_72h_v3.csv")
PBMC = pd.read_csv("C:\\Users\\DELL\\Desktop\\Shree_Tableau_files\\PBMC co-culture plate layout.csv")
```

```python
#Add timepoint column to each dataset
value_to_fill_24 = 24
value_to_fill_48 = 48
value_to_fill_72 = 72
new_column_name = 'Timepoint'
M_24.insert(0,'Timepoint',value_to_fill_24)
M_48.insert(0,'Timepoint',value_to_fill_48)
M_72.insert(0,'Timepoint',value_to_fill_72)
```

```python
Combined_data = pd.concat([M_24, M_48, M_72], ignore_index=True, sort=False)
```

```python
Combined_data.rename(columns={'Unnamed: 0': 'Mixture'},inplace = True)
```

```python
Combined_data.head()
```

```python
#Assigning well_id
Combined_data['Well_ID'] = Combined_data["Mixture"].str.split('_').str[-1].str.split('.').str[0]
```

```python
Combined_data.head()
```

```python
Combined_data = pd.merge(Combined_data, PBMC, on='Well_ID', how='outer', sort=False)
```

```python
Combined_data.head()
```

```python
Combined_data['Mixture'] = Combined_data['Mixture'].str[:20]
```
